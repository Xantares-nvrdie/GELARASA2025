```python
import pandas as pd

chunks = pd.read_csv('data/raw/sales.csv', parse_dates=['date'], chunksize=200000)
df = pd.concat(chunks)('../data/raw/sales.csv', parse_dates=['date'], chunksize=200000)
sample_list = []
for chunk in chunks:
    sample_list.append(chunk.sample(n=2000, random_state=42))

df_sample = pd.concat(sample_list)
df_sample.to_csv('../data/sample_small.csv', index=False)
print("sample_small.csv berhasil dibuat")
```
